```python
fruits= ['apple','banana','cherry','grapes','papaya']
print("original array:",fruits)
print("first element:",fruits[0])
```

```python
fruits.append('orange')
print("current array:",fruits)
```

```python
fruits.remove('grapes')
print("current array:",fruits)
print("Length of list:",len(fruits))
```

```python
points=(10, 20, 30,40,50)
print("Tuple:",points)
```

```python
print("third element:",points[2])
```

```python
rollno={1, 2, 4, 5, 2, 4}
print("set:",rollno)
```

```python
rollno.add(9)
print("set after adding:",rollno)
```

```python
rollno.remove(2)
print("set after removing:",rollno)
```

```python
a={4, 7, 9}
b={2, 4, 6}
print("union:",a|b)
print("intersection:",a&b)
print("difference:",a-b)
```

```python
student={
   'name': 'Alice',
    'rollno': 22,
    'dept': 'CSE'
}
print("Dictionary:", student)

student['grade'] = 'A'
print("After adding grade:", student)

del student['rollno']
print("After deleting rollno:", student)

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
print("Student name:",student['name'])
```

```python
keys = ['id', 'name', 'marks']
values = [522, 'Brahmani', 98]
student_info = dict(zip(keys, values))
print("Student Information:", student_info)
```
